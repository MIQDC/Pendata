## Fitur: **Petal Length**

### 1. Urutkan data berdasarkan Petal Length

| No | Petal Length | Class           |
| -- | ------------ | --------------- |

| 15 | 4.8          | Iris-versicolor |
| 8  | 4.9          | Iris-versicolor |
| 11 | 5.1          | Iris-versicolor |

Ambil threshold **Petal Length = 1.5**


### 2. Entropy Awal (sebelum split)

$$
p_1 = \frac{7}{15}, \quad p_2 = \frac{8}{15}
$$

$$
Entropy(D) = -\frac{7}{15}\log_2\frac{7}{15} - \frac{8}{15}\log_2\frac{8}{15} \approx 0.997
$$


### 3. Split Data:

* **Left (≤ 1.5)**: 7 data (No: 13, 3, 1, 2, 5, 4, 12) → semua *Iris-setosa*
* **Right (> 1.5)**: 8 data (No: 9, 7, 10, 6, 14, 15, 8, 11) → semua *Iris-versicolor*


### 4. Entropy Setelah Split

* Left: 7 setosa → Entropy = 0
* Right: 8 versicolor → Entropy = 0

$$
Entropy_{split} = \frac{7}{15} \cdot 0 + \frac{8}{15} \cdot 0 = 0
$$


### 5. Information Gain

$$
IG = 0.997 - 0 = \boxed{0.997}
$$


### 6. Gini Index

$$
Gini(D) = 1 - \left(\frac{7}{15}\right)^2 - \left(\frac{8}{15}\right)^2 = 1 - \frac{49}{225} - \frac{64}{225} = \frac{112}{225} \approx 0.498
$$

$$
Gini_{left} = 0, \quad Gini_{right} = 0
$$

$$
Gini_{split} = \boxed{0}
$$

## Menggunakan 15 data menampilkan visualisasi boundary dan decision tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree

# 1. Dataset 15 data
data = {
    'petal_length': [1.4, 1.4, 1.3, 1.5, 1.4, 4.7, 4.5, 4.9, 4.0, 4.6, 5.1, 1.5, 1.2, 4.7, 4.8],
    'petal_width':  [0.2, 0.2, 0.2, 0.2, 0.3, 1.4, 1.5, 1.5, 1.3, 1.5, 1.5, 0.2, 0.2, 1.6, 1.8],
    'class': [
        'Iris-setosa', 'Iris-setosa', 'Iris-setosa', 'Iris-setosa', 'Iris-setosa',
        'Iris-versicolor', 'Iris-versicolor', 'Iris-versicolor', 'Iris-versicolor',
        'Iris-versicolor', 'Iris-versicolor', 'Iris-setosa', 'Iris-setosa',
        'Iris-versicolor', 'Iris-versicolor'
    ]
}

# 2. DataFrame dan Encoding
df = pd.DataFrame(data)
df['label'] = df['class'].map({'Iris-setosa': 0, 'Iris-versicolor': 1})
X = df[['petal_length', 'petal_width']]
y = df['label']

# 3. Train Decision Tree
clf = DecisionTreeClassifier(criterion='entropy', max_depth=None)
clf.fit(X, y)

# 4. Visualisasi Decision Boundary
x_min, x_max = X['petal_length'].min() - 0.5, X['petal_length'].max() + 0.5
y_min, y_max = X['petal_width'].min() - 0.2, X['petal_width'].max() + 0.2

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)

# Plot data points
for label, color, name in zip([0, 1], ['blue', 'orange'], ['Iris-setosa', 'Iris-versicolor']):
    subset = df[df['label'] == label]
    plt.scatter(subset['petal_length'], subset['petal_width'],
                c=color, edgecolor='k', s=100, label=name)

plt.legend(title='Kelas', loc='upper left', markerscale=1.5)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Decision Boundary - Iris Setosa vs Versicolor (15 Data)')
plt.grid(True)
plt.show()

# 5. Print Rules
print("\nRules:")
print(export_text(clf, feature_names=list(X.columns)))

# 6. Visualisasi Decision Tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['petal_length', 'petal_width'],
          class_names=['setosa', 'versicolor'],
          filled=True, rounded=True)
plt.title("Decision Tree - 15 Data")
plt.show()
```

## Tambahkan 2 data baru dan visualisasi boundary dan decision tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text

# Data awal + 2 data baru
data = {
    'petal_length': [1.4, 1.4, 1.3, 1.5, 1.4, 4.7, 4.5, 4.9, 4.0, 4.6, 5.1, 1.5, 1.2, 4.7, 4.8, 1.6, 4.4],
    'petal_width':  [0.2, 0.2, 0.2, 0.2, 0.3, 1.4, 1.5, 1.5, 1.3, 1.5, 1.5, 0.2, 0.2, 1.6, 1.8, 0.3, 1.4],
    'label': [
        'Iris-setosa', 'Iris-setosa', 'Iris-setosa', 'Iris-setosa', 'Iris-setosa',
        'Iris-versicolor', 'Iris-versicolor', 'Iris-versicolor', 'Iris-versicolor',
        'Iris-versicolor', 'Iris-versicolor', 'Iris-setosa', 'Iris-setosa',
        'Iris-versicolor', 'Iris-versicolor',  # ini total 15
        'Iris-setosa', 'Iris-versicolor'       # tambahan 2
    ]
}

# Encode label jadi angka
df = pd.DataFrame(data)
df['label_num'] = df['label'].map({'Iris-setosa': 0, 'Iris-versicolor': 1})

# Model Decision Tree
X = df[['petal_length', 'petal_width']]
y = df['label_num']
clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
clf.fit(X, y)

# --- CETAK DATA BARU YANG DITAMBAHKAN ---
print(" 2 Data Baru yang Ditambahkan:")
print(df.iloc[15:17][['petal_length', 'petal_width', 'label']])
print()

# Plot decision boundary
x_min, x_max = X['petal_length'].min() - 0.5, X['petal_length'].max() + 0.5
y_min, y_max = X['petal_width'].min() - 0.5, X['petal_width'].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')

# Scatter plot data points
for label, color, name in zip([0, 1], ['blue', 'orange'], ['Iris-setosa', 'Iris-versicolor']):
    subset = df[df['label_num'] == label]
    plt.scatter(subset['petal_length'], subset['petal_width'],
                c=color, edgecolor='k', s=100, label=name)

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Decision Boundary Setelah Penambahan 2 Data Baru')
plt.legend(title='Kelas', loc='upper left')
plt.grid(True)
plt.show()

# Visualisasi pohon keputusan
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['petal_length', 'petal_width'],
          class_names=['setosa', 'versicolor'],
          filled=True, rounded=True)
plt.title("Decision Tree - 17 Data")
plt.show()

# Cetak rules
rules = export_text(clf, feature_names=['petal_length', 'petal_width'])
print(" Rules dari Pohon Keputusan:\n")
print(rules)
```

```python
def explain_rules(rules_text):
    print("\n Penjelasan Aturan Decision Tree:\n")
    lines = rules_text.strip().split("\n")
    indent_stack = []

    for line in lines:
        indent = line.count('|   ')
        content = line.strip()

        # Bersihkan rule
        rule_part = content.replace('|--- ', '')

        # Jika mencapai node daun
        if 'class:' in rule_part:
            cls = rule_part.split(':')[-1].strip()
            if indent_stack:
                kondisi = ' dan '.join(indent_stack[:indent])
                print(f"- Jika {kondisi}, maka prediksi: Iris-{ 'setosa' if cls == '0' else 'versicolor' }")
            else:
                print(f"- Prediksi langsung: Iris-{ 'setosa' if cls == '0' else 'versicolor' }")
        else:
            # Simpan kondisi ke stack
            if len(indent_stack) > indent:
                indent_stack = indent_stack[:indent]
            indent_stack.append(rule_part)

# Cetak penjelasan:
explain_rules(rules)
```
